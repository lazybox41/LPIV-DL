```python
features=data.drop(30, axis=1)
targets=data[30]

x_train, x_test, y_train, y_test=train_test_split(features,targets, test_size=0.2)
x_train=x_train[y_train==0]
x_test.shape
```

```python
class AutoEncoder(Model):
    def __init__(self, op):
        super().__init__()
        self.encoder=Sequential([
            Dense(16,activation='relu'),
            Dense(8,activation='relu')
        ])
        self.decoder=Sequential([
            Dense(16,activation='relu'),
            Dense(op,activation='sigmoid')
        ])
    def call(self,inputs):
        encoded=self.encoder(inputs)
        decoded=self.decoder(encoded)
        return decoded
```

```python
x_train=np.asarray(x_train).astype("float32")
x_test=np.asarray(x_test).astype("float32")

model = AutoEncoder(29)
model.compile(loss='mse', optimizer='adam')
h=model.fit(x_train, x_train, epochs=10)
```

```python
x_test
```

```python

rec=model.predict(x_train)
re=tf.keras.losses.mse(rec,x_train)
threshold=np.mean(re)+np.std(re)
preds=model.predict(x_test)
prede=tf.keras.losses.mse(preds,x_test)
prd=tf.math.less(threshold,prede)
```
