```python
import tensorflow as tf
import keras.backend as K
from tensorflow.keras.layers import Dense,Lambda,Embedding
import numpy as np
from keras.optimizers import SGD
from keras.models import Sequential
from keras.utils import to_categorical
```

```python
data = 'One morning as Gregor Samsa was waking up from his anxious dreams, he discovered that in his bed, he had been changed into a monsterous verminous bug. '
```

```python
words = data.split()
words
```

```python
words= [x.lower().replace(".","").replace(",","") for x in words]
words
```

```python
vocab=set(words)
vocab_size=len(vocab)
vocab_size
```

```python
cw = 2
ed = 100
```

```python
wi={word:i for i,word in enumerate(vocab)}
iw={i:word for i,word in enumerate(vocab)}
iw
```

```python
x_train=[]
y_train=[]

for index,i in enumerate(words):
    if (index <= cw-1) or (index >= len(words)-cw):
        continue
        
    s= index - cw
    e= index + cw + 1
    
    context = words[s:e]
    context.remove(i)
    target = i
    
    x_train.append([wi[w] for w in context])
    y_train.append([wi[i]])
```

```python
x_train = np.array(x_train)
y_train = np.array(y_train)
x_train
```

```python
y_train = to_categorical(y_train, num_classes=vocab_size)
```

```python
cbow=Sequential()
cbow.add(Embedding(input_dim=vocab_size, output_dim=ed, input_length=cw*2))
cbow.add(Lambda(lambda x: K.mean(x,axis=1),output_shape=(ed,)))
cbow.add(Dense(vocab_size,activation="softmax"))
```

```python
cbow.compile(loss='categorical_crossentropy', optimizer=SGD(), metrics=['accuracy'])
```

```python
h=cbow.fit(x_train, y_train, epochs=5000, batch_size=256)
```

```python
preds = cbow.predict(x_train)
```
